```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Loading dataset from UCI Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

print("Dataset Shape:", df.shape)
print("First 5 rows:", df.head())
print("Class Distribution:", df['quality'].value_counts())
```

```python
X = df.drop('quality', axis=1).values
y = (df['quality'] >= 6).astype(int).values  # Binary classification
#Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples:  {X_test.shape[0]}")
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(64, 32),   # Two hidden layers
    activation='tanh',             # NOT relu
    solver='lbfgs',                # NOT adam
    max_iter=1000,
    random_state=42
)
mlp.fit(X_train, y_train)
print("Trained")
predictions = mlp.predict(X_test) #predicting test data
```

```python
print("Classification Report")
print(classification_report(y_test, predictions, target_names=["Bad Quality", "Good Quality"]))

print("Confusion Matrix")
cm = confusion_matrix(y_test, predictions)
print(cm)
print("Confusion Matrix Breakdown:")
print(f"  True Negatives  (Bad correctly identified): {cm[0][0]}")
print(f"  False Positives (Bad wrongly as Good):      {cm[0][1]}")
print(f"  False Negatives (Good wrongly as Bad):      {cm[1][0]}")
print(f"  True Positives  (Good correctly identified):{cm[1][1]}")

accuracy = np.mean(predictions == y_test) * 100
print(f"Overall Accuracy: {accuracy:.2f}%")
```
